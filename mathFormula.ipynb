


Objective Function:
\begin{align*}
\text{obj} &= \sum_{b \in B} \sum_{t \in T} \text{gen\_cost}_b \cdot (\text{gen\_bus}_{b, t})^2 \\
&\quad + \text{peak\_price} \cdot \sum_{t \in T} \max(0, P_{t, 1} - P_{t, 2}) \\
&\quad + \text{offpeak\_price} \cdot \sum_{t \in T} \max(0, P_{t, 1} - P_{t, 2})
\end{align*}

Sets:
\begin{align*}
&\text{Set of time periods: } T \\
&\text{Set of buses: } B \\
&\text{Set of electric vehicles: } E
\end{align*}

Parameters:
\begin{align*}
&\text{PV production profiles: } PV_b, \text{ for all } b \in B \\
&\text{Demand profiles: } demand_b, \text{ for all } b \in B \\
&\text{Admittance matrix: } Y_{\text{bus}_{b_1, b_2}}, \text{ for all } b_1, b_2 \in B \\
&\text{Admittance imaginary matrix: } B_{\text{bus}_{b_1, b_2}}, \text{ for all } b_1, b_2 \in B \\
&\text{Generation costs: } \text{gen\_cost}_b, \text{ for all } b \in B \\
&\text{Maximum generation limits: } \text{gen\_max}_b, \text{ for all } b \in B \\
&\text{Number of electric vehicles: } \text{num\_EVs} \\
&\text{Charging limit for EVs: } \text{charge\_limit} \\
&\text{Discharging limit for EVs: } \text{discharge\_limit} \\
&\text{Start time of peak hours: } \text{peak\_start\_time} \\
&\text{End time of peak hours: } \text{peak\_end\_time} \\
&\text{Peak price: } \text{peak\_price} \\
&\text{Off-peak price: } \text{offpeak\_price}
\end{align*}

Variables:
\begin{align*}
&\text{Power injection at each bus: } P_{t, b}, \text{ for all } t \in T, b \in B \\
&\text{Voltage angles at each bus: } \theta_{b, t}, \text{ for all } b \in B, t \in T \\
&\text{Generator power output for each bus: } \text{gen\_bus}_{b, t}, \text{ for all } b \in B, t \in T \\
&\text{EV charging/discharging power: } \text{EV\_charge}_{e, t}, \text{ for all } e \in E, t \in T
\end{align*}

Constraints:
\begin{align*}
&\text{Power flow constraints:} \\
&\quad \theta_{0, t} = 0, \quad \forall t \in T \\
&\quad P = Y_{\text{bus}} \cdot \theta \\
&\quad P_{t, b} = \text{demand}_b - \text{PV}_b - \text{gen\_bus}_{b, t}, \quad \forall b \in B, \forall t \in T \\
&\text{Generator capacity constraints:} \\
&\quad \text{gen\_bus}_{b, t} \leq \text{gen\_max}_b, \quad \forall b \in B, \forall t \in T \\
&\text{EV charging and discharging constraints:} \\
&\quad \text{EV\_charge}_{e, t} \geq 0, \quad \text{if initial\_SOC}_e < \text{charge\_limit}, \quad \forall e \in E, \forall t \in T \\
&\quad \text{EV\_charge}_{e, t} \leq 0, \quad \text{if } \text{peak\_start\_time} \leq t \leq \text{peak\_end\_time} \text{ and } \text{initial\_SOC}_e > \text{discharge\_limit}, \quad \forall e \in E, \forall t \in T \\
&\text{Energy neutrality constraint:} \\
&\quad \sum_{e \in E} \text{EV\_charge}_{e, t} = P_{t, 1} - P_{t, 2}, \quad \forall t \in T
\end{align*}



#### Parameters

The parameters used in the problem are defined as follows:

- PV production profiles: $PV_b$, for all $b \in B$
- Demand profiles: $\text{demand}_b$, for all $b \in B$
- Admittance matrix: $Y_{\text{bus}_{b_1, b_2}}$, for all $b_1, b_2 \in B$
- Admittance imaginary matrix: $B_{\text{bus}_{b_1, b_2}}$, for all $b_1, b_2 \in B$
- Generation costs: $\text{gen\_cost}_b$, for all $b \in B$
- Maximum generation limits: $\text{gen\_max}_b$, for all $b \in B$
- Number of electric vehicles: $\text{num\_EVs}$
- Charging limit for EVs: $\text{charge\_limit}$
- Discharging limit for EVs: $\text{discharge\_limit}$
- Start time of peak hours: $\text{peak\_start\_time}$
- End time of peak hours: $\text{peak\_end\_time}$
- Peak price: $\text{peak\_price}$
- Off-peak price: $\text{offpeak\_price}$

#### Variables

The variables used in the problem are defined as follows:

- Power injection at each bus: $P_{t, b}$, for all $t \in T$, $b \in B$
- Voltage angles at each bus: $\theta_{b, t}$, for all $b \in B$, $t \in T$
- Generator power output for each bus: $\text{gen\_bus}_{b, t}$, for all $b \in B$, $t \in T$
- EV charging/discharging power: $\text{EV\_charge}_{e, t}$, for all $e \in E$, $t \in T$

### Sets

The sets used in the problem are defined as follows:

- Set of time periods: $T$
- Set of buses: $B$
- Set of electric vehicles: $E$

#### Objective Function

The objective function is defined as follows:

$$
\text{obj} = \sum_{b \in B} \sum_{t \in T} \text{gen\_cost}_b \cdot (\text{gen\_bus}_{b, t})^2 
$$

#### Constraints

The constraints of the problem are defined as follows:

- Power flow constraints:
  - $\theta_{0, t} = 0$, for all $t \in T$
  - $P = -Y_{\text{bus}} \cdot \theta$
  - $P_{t, b} = \text{demand}_b - \text{PV}_b - \text{gen\_bus}_{b, t}$, for all $b \in B$, $t \in T$

- Generator capacity constraints:
  - $\text{0}\leq\text{gen\_bus}_{b, t} \leq \text{gen\_max}_b$, for all $b \in B$, $t \in T$

- EV charging and discharging constraints:
  - $\text{EV\_charge}_{e, t} \geq 0$, if $\text{initial\_SOC}_e < \text{charge\_limit}$, for all $e \in E$, $t \in T$
  - $\text{EV\_charge}_{e, t} \leq 0$, if $\text{peak\_start\_time} \leq t \leq \text{peak\_end\_time}$ and $\text{initial\_SOC}_e > \text{discharge\_limit}$, for all $e \in E$, $t \in T$

- Energy neutrality constraint:
  - $\sum_{e \in E} \text{EV\_charge}_{e, t} = P_{t, 1} - P_{t, 2}$, for all $t \in T$
